### 27. Monthly_Income sütunundaki değerleri belirtilen gibi değiştirin ve 10 tane random veri seçin. ( Not: Bu sütun çıktısının farklılık göstermesi önemli değildir.)
No Income = 0, 
Below Rs.10000 = 7500, 
10001 to 25000 = 17500, 
25001 to 50000 = 37500, 
More than 50000 = 75000


```python
# alternatif çözüm
# df["Monthly_Income"] = df.Monthly_Income.map({"No Income": 0, "Below Rs.10000": 7500, "10001 to 25000": 17500, 
#                                             "25001 to 50000": 37500, "More than 50000": 75000})
```

```python
df.Monthly_Income.loc[np.where(df.Monthly_Income == "No Income")] = 0
df.Monthly_Income.loc[np.where(df.Monthly_Income == "Below Rs.10000")] = 7500
df.Monthly_Income.loc[np.where(df.Monthly_Income == "10001 to 25000")] = 17500
df.Monthly_Income.loc[np.where(df.Monthly_Income == "25001 to 50000")] = 37500
df.Monthly_Income.loc[np.where(df.Monthly_Income == "More than 50000")] = 75000
df.sample(20)
```

### 28. Monthly_Income sütunun tipini int32'ye çeviriniz.

```python
df["Monthly_Income"] = df.Monthly_Income.astype('int32')
df.info()
```

## 31. cut() fonksiyonu ile Monthly_Income verisine göre belirtilen şekilde kategorik sınıflandırma yaparak "Income_Cat" olarak yeni sütun ekleyin.
0 - 7500(Haric) = No
7500- 17500(Haric) = Low
17501 - 37500(Haric) = Medium
37500- 75000(Dahil) = High

```python
bins = [-1,7499, 17499, 37499, 75001]

group_names = ["No", "Low", "Medium", "High"]

df["Income_Cat"] = pd.cut(df["Monthly_Income"], bins, labels=group_names)

df.head()
```
