## 2: Plot 12 most dangerous cities

```python
cities = kill.city.value_counts().head(12) # kill.city.value_counts().sort_values(ascending=False).head(12)
cities
```

```python
# top 12 the most dangerous cities
```

```python
cities = kill.city.value_counts().head(12) # kill.city.value_counts().sort_values(ascending=False).head(12)
incident_counts = np.array(cities.values)
incident_counts
```

```python
#array
```

```python
cities.index
```

```python
#index
```

```python
fig, ax = plt.subplots(figsize= (10, 5))

plt.title(label= "The Most Dangerous Cities", color="red")
plt.xticks(rotation=45, color="blue")

sns.barplot(data=kill, x=cities.index, y=incident_counts)

plt.tight_layout();
```

```python
# with seaborn
```

```python
fig, ax = plt.subplots(figsize= (10, 5))

plt.title(label= "The Most Dangerous Cities", color="blue",fontsize=20)
plt.xticks(rotation=45)
order = cities.index

sns.countplot(data=kill, x="city", order=order);
#sns.countplot(data=kill[kill['city'].isin(order)], x="city", order=order);

plt.tight_layout()
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states

```python
kill.head()
```

```python
kill.state.value_counts().head(20)
```

```python
#the 20 most dangerous states
```

```python
top_20_states = kill.state.value_counts().head(20)
top_20_states
```

```python
fig, ax = plt.subplots(figsize= (10, 5))

plt.title(label= "Most Dangerous States", color="blue")
plt.xticks(rotation=45, color="blue")

sns.barplot(data=kill, x=top_20_states.index, y=top_20_states.values, palette="tab10");
```

```python
# with seaborn barplot
```

```python
fig, ax = plt.subplots(figsize= (10, 5))

plt.title(label= "Most Dangerous States", fontsize=20, color="blue")
plt.xticks(rotation=45, color="blue")

order = top_20_states.index

sns.countplot(data=kill, x="state", order=order);
#sns.countplot(data=kill[kill['state'].isin(order)], x="state", order=order);
```

## 6: Plot number of Flee Types

```python
flee_types = kill["flee"].value_counts()
flee_types
```

```python
#flee types
```

```python
sns.barplot(data=kill, x = flee_types.index, y=flee_types.values)
```

```python
ax = sns.barplot(data=kill, x = flee_types.index, y=flee_types.values)

ax.bar_label(ax.containers[0]) # shows values on the bars

plt.show()
```

```python
# with seaborn
```

```python
ax = sns.countplot(data=kill, x="flee", hue = "gender")

for j in [0,1]:
        ax.bar_label(ax.containers[j])
        
#ax.bar_label(ax.containers[0])
#ax.bar_label(ax.containers[1])   bu şekilde de çalışır

plt.show()
```
